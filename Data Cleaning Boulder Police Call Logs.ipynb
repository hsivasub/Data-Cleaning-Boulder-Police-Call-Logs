### Here the columns are not correct. For the first few rows, the Response Date is present under the Incident_ID column. Also for the first few columns, the CaseNumber and Problem columns are interchanged.

### Interchanging the Response Date and Incident ID at specific rows

```python
df['ResponseDate'], df['Incident_ID'] = np.where(df['ResponseDate']=='1', [df['Incident_ID'], df['ResponseDate']], [df['ResponseDate'], df['Incident_ID'] ])
df['ResponseDate'], df['Incident_ID'] = np.where(df['ResponseDate']=='2', [df['Incident_ID'], df['ResponseDate']], [df['ResponseDate'], df['Incident_ID'] ])
df['ResponseDate'], df['Incident_ID'] = np.where(df['ResponseDate']=='3', [df['Incident_ID'], df['ResponseDate']], [df['ResponseDate'], df['Incident_ID'] ])
df['ResponseDate'], df['Incident_ID'] = np.where(df['ResponseDate']=='4', [df['Incident_ID'], df['ResponseDate']], [df['ResponseDate'], df['Incident_ID'] ])
```

```python
df.tail()
    
```

```python
df.ResponseDate.value_counts()
```

### Now the Problem and Case Number columns are interchanged

```python
df.Problem.value_counts()
```

### One important pattern found from this data is that, all the case numbers start with the number 2. We will use that to interchange the column

```python
df['Problem'],df['CaseNumber'] =np.where(df['Problem'].astype(str).str[0]=='2', [df['CaseNumber'],df['Problem']]  ,[df['Problem'],df['CaseNumber']]  )
```

```python
df.head()
```

```python
df.Problem= df.Problem.combine_first(df.CaseNumber)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.Problem.value_counts()
```

### Now the problem column is perfect

```python
df.CaseNumber.value_counts()
```

### Note here that the Case Number column has some Problems, so we will convert them as NaN

```python
df['CaseNumber']= np.where(df['CaseNumber'].astype(str).str[0]!='2', 'NaN', df['CaseNumber'])
```

### Here the Column "CaseNumber" has 34% Null values. Since Case Number does not have any significance, we wont worry about them now

```python
df.head()
```

```python
df[['A','B']]=df.Problem.str.split('-',expand=True)
```

```python
df.head()
```

```python
def status(df):
    if 'I/P' in df:
        return "I/P"
    elif 'J/O' in df:
        return "J/O"
    else: 
        return "Complete"
        
```

```python
df['Status']= df['B'].apply(status)
```

```python
df.head()
```

```python
df['Status'].value_counts()
```

```python
df.tail()
```

```python
def crime(df):
    if 'J/O' in df:
        return(df[:-3])
    elif 'I/P' in df:
        return(df[:-3])
    else:
        return(df)
```

```python
df['Crime']= df['B'].apply(crime)
```

```python
df.head()    
```

```python
df.tail()
```

```python
def removec(df):
    if df.endswith(' C'):
        return(df[:-1])
    else:
        return(df)
```

```python
df['Crime']=df['Crime'].apply(removec)
```

```python
df['Crime'].value_counts()
```

```python
df=df.drop(['A', 'B'], axis=1)
df.head()
```

```python
df['Time']=df['ResponseDate'].str[-8:]
df.head()
```

```python
df['ResponseDate']=df['ResponseDate'].str[:-8]
df.head()
```

```python
df['Time'].value_counts()
```

```python
am=0
pm=0
for i in df['Time']:
    if 'AM' in i:
        am+=1
    elif 'PM' in i:
        pm+=1
```

```python
print(am+pm)
```

### If we remove all the 'AM' and 'PM' rows, then there will be a lose of about 10% of data. Thus we will convert them as Unknown

```python
def timeconv(df):
    if 'AM' in df:
        return 'Unknown'
    elif 'PM' in df:
        return 'Unknown'
    else:
        return df
```

```python
df.tail()
```

```python
df.Crime.value_counts()
```

```python
df['Time']=df['Time'].apply(timeconv)
```

```python
df.tail()
```

```python
df['ResponseDate']=df['ResponseDate'].str[:10]
```

```python
df['ResponseDate'].value_counts()
```

```python
def clean(df):
    if df.find(' ')>0:
        return(df[:9])
    else:
        return(df)
```

```python
df['ResponseDate']=df['ResponseDate'].apply(clean)
```

```python
df['ResponseDate'].value_counts()
```
